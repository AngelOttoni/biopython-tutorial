# **Index of each nucleotide of a DNA sequence**


**Pergunta Biológica 1:**
- Quão importante é encontrar o TATA Box em regiões promotoras e a quais posições correspondem tais nucleotídeos da seguinte sequência de DNA?

**Resposta:**
- O TATA Box ou caixa de Goldberg-Hogness é uma sequência de DNA encontrada na região central do promotor de genes em arquéias e eucariotos onde liga-se a proteína de ligação ao TATA (TBP) e outros fatores de transcrição.

- A transcrição do gene pela RNA polimerase II depende da regulação do promotor central por elementos reguladores de longo alcance, como potenciadores e silenciadores. Sem a regulação adequada da transcrição, os organismos eucarióticos não são capazes de responder adequadamente ao seu ambiente.

- Mutações no TATA Box podem levar a:
  - Diminuição da transcrição (menos expressão da proteína correspondente);
  - Ausência da transcrição esperada.

- **Por exemplo:** Se numa amostra sequenciada de um paciente espera-se encontrar o TATA Box na região promotora do gene de interesse, podemos suspeitar de um processo de mutação o qual está associado a algumas doenças em humanos que podem ser:
  - Câncer gástrico;
  - Ataxia espinocerebelar;
  - Doença de Huntington;
  - Cegueira;
  - β-talassemia;
  - Imunossupressão;
  - Síndrome de Gilbert;
  - HIV-1.

# **Calculation of GC % content of a given DNA/RNA sequence**

**Pergunta Biológica:**
- É relevante saber o % de GC de uma amostra?

**Resposta:** SIM!

**Exemplo 1:**
- Para calcular a Temperatura de Melting para uma PCR: Quanto maior o % de GC, mais energia precisa ser fornecida para poder quebrar as ligações triplas de Hidrogênio formadas entre G e C.

400px-Interacao_das_bases_nitrogenadas.png

**Exemplo 2:**
- O conteúdo médio de GC varia de organismo para organismo:
  - Humanos: de 35% a 60% de GC;
  - *Saccharomyces cerevisiae:* ~ 38% de GC;
  - *Arabidopsis thaliana:* ~ 36% de GC;
  - *Plasmodium falciparum:*: ~ 20% de GC (rico em AT);
  - Micobactérias: 64 a 70% de GC.
- Como observado anteriormente, em geral as micobactérias são os organismos que apresentam maior percentual de GC. Contabilizando o % de GC de uma amostra, podemos suspeitar que haja presença de espécies do gênero *Mycobacterium*.

- The `Bio.SeqUtils` module in Biopython contains the function `gc_fraction`, which is a replacement for the deprecated `GC` function.

- The `gc_fraction` function is used to calculate the GC (guanine-cytosine) content in a DNA or RNA sequence.

- It calculates the fraction of bases that are either guanine (G) or cytosine (C) in the sequence.

````python
# Criando a variável para receber uma porção de um gene de Mtb (Mycobacteirum tuberculosis H37Rv)...
my_seq = Seq("GTCATATGAAATACCCCGTCGAGGGCGGCGGAAACCAGGACTGGTGGCCCAACCGGCTCAATCTGAAGGTACTGCACCAAAACCCGGCCGTCGCTGACCCGATGGGTGCGGCGTTCGACTATGCCGCGGAGGTCGCGACCATCGACGTTGACGCCCTGACGCGGGACATCGAGGAAGTGATGACCACCTCGCAGCCGTGGTGGCCCGCCGACTACGGCCACTACGGGCCGCTGTTTATCCGGATGGCGTGGCACGCTGCCGGCACCTACCGCATCCACGACGGCCGCGGCGGCGCCGGGGGCGGCATGCAGCGGTTCGCGCCGCTTAACAGCTGGCCCGACAACGCCAGCTTGGACAAGGCGCGCCGGCTGCTGTGGCCGGTCAAGAAGAAGTACGGCAAGAAGCTCTCATGGGCGGACCTGATTGTTTTCGCCGGCAACTGCGCGCTGGAATCGATGGGCTTCAAGACGTTCGGGTTCGGCTTCGGCCGGGTCGACCAGTGGGAGCCCGATGAGGTCTATTGGGGCAAGGAAGCCACC")
my_seq
````

````python
# Calculando o % GC manualmente...

gc_percent = 100 * float(my_seq.count("G") + my_seq.count("C")) / len(my_seq)

print(f"GC content calculated manually: {gc_percent:.2f}%")
````

````python
# Calculando o % GC com o módulo Bio.SeqUtils...

from Bio.SeqUtils import gc_fraction
````

````python
gc_content = gc_fraction(my_seq)

print(f"GC content using gc_fraction: {gc_content:.2f}%")
````

# **Tradução**

## **ORF vs. CDS**

**IMPORTANTE:**
- Uma **ORF (*Open Reading Frame*)** é uma sequência de DNA que começa com o códon de início “ATG” (nem sempre) e termina com qualquer um dos três códons de terminação (TAA, TAG, TGA). A ORF é geralmente prevista com base na sequência de DNA e não foi comprovada a transcrição;
- Dependendo do ponto de partida, existem 6 maneiras possíveis (3 na fita direta ou parental e 3 na fita complementar) de traduzir qualquer sequência de nucleotídeos em sequência de aminoácidos de acordo com o código genético. Essas são chamadas de **frames de leitura**.
- Uma **CDS (*Coding Sequence*)** é a região real do DNA que é traduzida para formar proteínas (sem os íntros, no caso de eucariotos), em outras palavras uma CDS se refere àqueles nucleotídeos (exons concatenados) que podem ser divididos em códons que são traduzidos em aminoácidos pelo mecanismo de tradução ribossomal.

## **Start-codon**

**IMPORTANTE:**
- O códon de início é o primeiro códon de um transcrito de RNA mensageiro (mRNA) traduzido por um ribossomo;
- O códon inicial sempre codifica para Metionina (Met) em eucariotos e Archaea e uma N-formilmetionina (fMet) em bactérias, mitocôndrias e plastídeos;
- O códon de início mais comum é AUG (isto é, ATG na sequência de DNA correspondente);
- Os códons de início alternativos são diferentes do códon AUG padrão e são encontrados tanto em procariotos (bactérias e arquéias) quanto em eucariotos;
- Os códons de início alternativo ainda são traduzidos como Met quando estão no início de uma proteína (mesmo se o códon codificar um aminoácido diferente de outra forma).

**Eucariotos:**
- Códons de início alternativos (não-AUG) são muito raros em genomas eucarióticos;
- Códons de início não AUG de ocorrência natural foram relatados para alguns mRNAs celulares;
- Sete das nove possíveis substituições de nucleotídeo único no códon inicial AUG da diidrofolato redutase foram funcionais como locais de início da tradução em células de mamíferos;
- Além da via canônica do códon Met-tRNA Met e AUG, as células de mamíferos podem iniciar a tradução com valina usando um valil-tRNA específico que decodifica o códon CUG.
- *Candida albicans* usa um códon de início CAG.

**Procariotos:**
- Os procariotos usam códons de início alternativo, principalmente GUG e UUG.
- 83% dos CDCs de *E. coli* usam AUG como códon de início, 14% usam GUG e 3% usam AUU e CUG. Quando algum destes códons está sendo usado como códon de início, a tradução sempre será para Metionina, embora usualmente seja traduzido para outro aminoácido, exemplo: AUU normalmente traduz para uma Isoleucina, mas como códon de início não-estandar, traduz para uma Metionina;
- As regiões de codificação bem conhecidas que não têm códons de iniciação AUG são aquelas de lacI (GUG) e lacA (UUG) no operon lac de *E. coli*.

**Mitocôndrias:**
- Os genomas mitocondriais usam códons de início alternativos de forma mais significativa (AUA e AUU em humanos).

**Referência:**

## **Mudando os nucleotídeos de uma sequência**

**IMPORTANTE:**
- As letras de uma sequência de um objeto Seq são imutáveis:

````python
# Criando a variável...
my_seq = Seq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
my_seq
````

````python
# Tentando trocar uma letra da sequência anterior...
my_seq[5] = "G"
````

Porém, isto pode mudar, usando o seguinte método:

````python
# Importando a MutableSeq
from Bio.Seq import MutableSeq
print("Setup ok!")
````

````python
# Trocando uma letra da sequência anterior sem obter erro...
mutable_seq = MutableSeq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
mutable_seq
````

````python
# Trocando um "C" por um "" na quinta posição da sequência anterior...
mutable_seq[5] = "C"
mutable_seq
````

Sequência original: GCCAT   **T**   GTAATGGGCCGCTGAAAGGGTGCCCGA

Sequência modificada: GCCAT   **C**   GTAATGGGCCGCTGAAAGGGTGCCCGA

````python
# Removendo o primeiro "T" localizado na quarta posição...
mutable_seq.remove("T")
mutable_seq
````

Sequência modificada 1: GCCA   **T**   CGTAATGGGCCGCTGAAAGGGTGCCCGA

Sequência modificada 2: GCCACGTAATGGGCCGCTGAAAGGGTGCCCGA

````python
# Invertendo a sequência anterior...
mutable_seq.reverse()
mutable_seq
````

Sequência modificada 2: GCCACGTAATGGGCCGCTGAAAGGGTGCCCGA

Sequência modificada 3: AGCCCGTGGGAAAGTCGCCGGGTAATGCACCG

````python
# Voltando a criar um objeto "Seq"...
new_seq = mutable_seq.toseq()
new_seq
````

## **Nucleotídeos desconhecidos de uma sequência**

**Problema Biológico:**

- É comum obter letras desconhecidas após sequenciamento de DNA ou proteínas.
- Para sequências de DNA ou RNA, nucleotídeos desconhecidos são comumente denotados pela letra “N”, enquanto que para proteínas, “X” é comumente usado para aminoácidos desconhecidos.
- O objeto UnknownSeq é uma subclasse do objeto Seq básico e seu objetivo é representar uma sequência em que descobramos o seu tamanho, mas não as letras reais que a compõem.
- Pode ser usado um objeto Seq normal nesta situação, mas gasta muita memória para armazenar uma string de um milhão de caracteres “N” quando você poderia simplesmente armazenar uma única letra “N” e o comprimento desejado como um inteiro .
- Alguns formatos de arquivo de sequência nem sempre incluem a sequência real, por exemplo, os arquivos GenBank e EMBL podem incluir uma lista de recursos, mas para a sequência apenas apresenta as informações contig.

**Formatos de arquivos:**

**FASTA vs. QUAL:**

- Os **arquivos QUAL** usados no trabalho de sequenciamento possuem pontuações de qualidade, mas nunca contêm uma sequência - em vez disso, está o **arquivo FASTA** que possui a sequência;
- O formato de arquivo QUAL é um formato adicional relacionado ao FASTA.
Um arquivo FASTA às vezes é acompanhado por um arquivo QUAL, principalmente quando o arquivo fasta contém sequências geradas em um instrumento de sequenciamento de alto rendimento;
- Os arquivos QUAL armazenam uma pontuação de qualidade Phred (inteiro não negativo) para cada base em uma sequência armazenada no formato FASTA.

**Exemplo de arquivo FASTA:** Possui a sequência de nucleotídeos
```
>seq1 db-accession-149855
CGATGTC
>seq2 db-accession-34989
CATCG
```

**Exemplo de arquivo QUAL:** Possui as pontuações de qualidade do processo de sequenciamento
```
>seq1 db-accession-149855
40 39 39 4 50 1 100
>seq2 db-accession-34989
3 3 10 42 80
```

**Referência:**

http://scikit-bio.org/docs/0.2.2/generated/skbio.io.fasta.html

````python
# Importando o Módulo UnKnownSeq
from Bio.Seq import UnknownSeq
print("Setup ok!")
````

````python
# Criando uma sequência contendo "N"...
unknown_dna = UnknownSeq(20, character = "N")
print(unknown_dna)
````

````python
# Obtendo o tamanho e o tipo de caracter da sequência proteica obtida a partir da sequência "unknown_dna"...
unknown_protein = unknown_dna.translate()
unknown_protein
````
